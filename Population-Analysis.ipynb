```python
import csv
f=open('age.csv')
data=csv.reader(f)
result=[]

for row in data:
    if '서울특별시 송파구 잠실3동' in row[0]:
        for i in row[1::39]:
            result.append(int(i))

import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.bar(range(2018,2024),result)
plt.title('연도별 잠실 3동의 총인구수')
plt.xlabel('연도')
plt.ylabel('인구')
plt.show()
```

```python
import csv
f=open('age.csv')
data=csv.reader(f)
result1 = []
result2 = []

for row in data:
    if '서울특별시 송파구 잠실3동' in row[0]:
        for i in row[4::39]:
            result1.append(int(i))
        for i in row[5::39]:
            result2.append(int(i))

result3 = []
for i in range(len(result1)):
    result3.append(result1[i] + result2[i])
print(result3)

import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.bar(range(2018,2024),result3)
plt.title('연도별 잠실 3동의 20대~30대 인구 그래프')
plt.xlabel('연도')
plt.ylabel('인구')
plt.show()
```
